```python
import numpy 
import pandas as pd
import matplotlib
from sklearn import preprocessing

file_path = "/Users/vikasramaneti/workspace/CS677/P2-nv-p2/src/main/java/edu/usfca/cs/mr/outputs/solar_wind_result.csv"
def load_solar_wind_data(file_path=file_path):
    return pd.read_csv(file_path, sep="\s+", names=["GeoHash", "Avg Temp", "Avg Rainfall", "Avg Solar Radiation", "Avg WindSpeed"])



solar_wind = load_solar_wind_data()
solar_wind


```

```python
solar_wind_sub = solar_wind.loc[:, "Avg Temp": "Avg WindSpeed"]
x = solar_wind_sub.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df.columns = ['Normalized Avg Temp', "Normalized Avg Rainfall", "Normalized Avg Solar Radiation", "Normalized Avg WindSpeed"]
df
```

```python
final_wind_solar = pd.concat([solar_wind, df], axis=1)
final_wind_solar.head(3)
```

```python
final_wind_solar["Solar Score"] = (5*final_wind_solar["Normalized Avg Solar Radiation"]) + (-1*final_wind_solar["Normalized Avg Temp"]) + (1*final_wind_solar["Normalized Avg Rainfall"]) 
final_wind_solar["Wind Score"] = (5*final_wind_solar["Normalized Avg WindSpeed"]) + (-1*final_wind_solar["Normalized Avg Temp"]) + (1*final_wind_solar["Normalized Avg Rainfall"])
final_wind_solar["Solar Wind Score"] = final_wind_solar["Solar Score"] + final_wind_solar["Wind Score"]

final_wind_solar.head(3)
```

```python
output_solar_score = final_wind_solar.sort_values("Solar Score", ascending=False)
output_solar_score.head(10)
```

```python
output_wind_score = final_wind_solar.sort_values("Wind Score", ascending=False)
output_wind_score.head(10)
```

```python
output_solar_wind_score = final_wind_solar.sort_values("Solar Wind Score", ascending=False)
output_solar_wind_score.head(10)
```
